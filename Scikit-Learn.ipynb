```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
# load breast cancer dataset from sklearn
brst_cncr = datasets.load_breast_cancer() 
# convert the dataset to a dataframe
brst_cncr_df = pd.DataFrame(data = brst_cncr.data, columns = brst_cncr.feature_names)
brst_cncr_df['target'] = pd.Series(brst_cncr.target)
brst_cncr_df
```

```python
# check whether the data set is balanced or not
negative_exaple_count = brst_cncr_df['target'][brst_cncr_df['target'] == 0].count()
positive_exaple_count = brst_cncr_df['target'][brst_cncr_df['target'] == 1].count()
print(f'# of positive examples is {positive_exaple_count}, and # of negative examples is {negative_exaple_count}!')
```

```python
from sklearn.model_selection import train_test_split

X = brst_cncr.data
y = brst_cncr.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn import svm
model = svm.SVC(gamma=0.000001, C=1)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
model.score = f1
model.score
```

```python
# change the kernel from rbf to linear
model.set_params(kernel='linear').fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
model.score = f1
model.score
```
